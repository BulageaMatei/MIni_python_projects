```python
import random

def extract_info(dict):
    """Extracts and returns relevant information from a person dictionary."""
    name = dict["name"]
    description = dict["description"]
    country = dict["country"]
    followers = dict['follower_count']
    return name, description, country, followers

def main():    
    game_over = False
    score = 0
    first_random_pick = random.choice(data)
    
    while game_over == False:
        second_random_pick = random.choice(data)
        name_A, description_A, country_A, followers_A = extract_info(first_random_pick)
        name_B, description_B, country_B, followers_B = extract_info(second_random_pick)
        
        print(f"Compare A: {name_A}, a {description_A}, from {country_A}")
        print("VS")
        print(f"Compare B: {name_B}, a {description_B}, from {country_B}")
            
        choice = input("Who has more followers? Type 'A' or 'B':").upper()    
    
        if (choice == 'A' and followers_A > followers_B) or (choice == 'B' and followers_B > followers_A):
            print("You are right!")
            first_random_pick = second_random_pick
            score += 1
            
            
        else:
            print(f"Sorry, that's wrong! Final score {score}")
            game_over = True

main()
```
